# DS-SF-30 | Unit Project 1 - Grand Theft Auto

## Evaluate the following Problem Statement

Using Jan 2016 - Dec 2016 data from the government website - https://data.sfgov.org - determine the likelihood of theft from locked cars in a particular San Francisco district by analyzing incident reports (category, Day Of Week, Date, Time, Police Dept District, Incident Address, location) filed to the police.


## Risks and Assumptions for the Data used

Risks:
* Some reports to the police may be fraudulent.
* The dataset maybe missing data (i.e. not complete).
* Incident Reports maybe misclassified.

Assumptions:
* All vehicle thefts have been reported.
* Vehicle thefts have been reported timely.
* Each incident report is for one vehicle.


##  Variables of the dataset, including the response and predictors

Incident Number, Crime Category, Day Of Week, Date, Time, Police District, Address, location.

Possible predictor variables | Possible response variables
---| ---
Time | Higher number of Incidences during early morning
Day of Week | Higher number of Incidences during week
Location | Higher number of Incidences in certain districts


## Data dictionary with classification of available variables

Variable | Description | Type of Variable
---| ---| ---
Incident Number |Report ID generated by police dept  | continuous
Crime Category | Type of Crime | categorical
Day Of Week | Day of the week the incident occurred | categorical
Date | Incident Date | continuous
Time | Incident Time | continuous
Police District | Police District ID | categorical
Address | Address of where the incident occurred | continuous
location | Coordinate location of the incident | continuous



## Exploratory data analysis methods

Possible methods:
* Histograms
* Mean plot
* Box plot

These methods help explore the spread of the date, highlight any outliers and help predict the likelihood of theft in a location in San Francisco.
